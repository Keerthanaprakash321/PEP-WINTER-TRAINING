```python
lst1= [1,3,4,56]
lst2 = lst1.copy()
lst2[1]=1000
lst2,lst1
```

### shallow copy nested list

```python
list1= [[1,2,3,545],[23,68.9]]
list2 = list1.copy()
list1[1][0]=100
list1
list2
```

```python
#deep copy
import copy
list1  =[1,2,3,4]
list2 =copy.deepcopy(list1)
list1[1]=100
list1
list2
```
