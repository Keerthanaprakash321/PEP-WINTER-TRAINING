##fibonacci 22/1/2026

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)


n = int(input())


for i in range(n):
    print(fibonacci(i), end=" ")
```

## swap case


```python
text = "hello world"
print(text.swapcase())
```

```python
def swap(ch):
    result=""
    for i in ch:
        if i.isupper():
            result+=i.lower()
        elif i.islower():
            result+=i.upper()
        else:
            result+=i
    return result
s=input()
convert=swap(s)
print(convert)
```
