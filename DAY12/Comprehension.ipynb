## Comprehension in list

```python
names=["keerthh","ash","vyas","tdgtf"]
names2 = [n.title() for n in names if len(n)==4]
print(names2)
```

## Dictionary comprehension

```python
inventory = {"movie":"siisk","cast":"jhdjs","revenue":"dsff"}
inventory2 = {n.title() for n in inventory if len(n)==5}
print(inventory2)
#not correct
```

```python
numbers = [1, 2, 3, 4]
squares = {n: n*n for n in numbers}
print(squares)
```
